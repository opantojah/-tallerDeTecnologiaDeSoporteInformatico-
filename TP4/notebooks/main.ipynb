```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import numpy as np

data = pd.read_csv("../data/dnrpa-robos-recuperos-autos-202103.csv")
# data = data.loc[0:100]

x = data['automotor_marca_descripcion'].unique()
y = data['automotor_marca_descripcion'].value_counts().tolist()  # lo que estoy contando seria la cantidad de robos que tiene cada marca
y.reverse() # muestre de menor a mayor

plt1.bar(x,y)
plt1.title('Robos segun marcas')
plt1.xlabel('marcas')
plt1.ylabel('cantidad de robos y recuperos')
plt1.xticks(rotation=90,fontsize=12)
plt1.show()

x2 = data['registro_seccional_provincia'].unique()
nombresX2 = ['BsAs','CABA','CÃ³rdoba','Mendoza']
x2 = np.array(nombresX2)
y2 = data['registro_seccional_provincia'].value_counts().tolist()
y2.reverse()

plt2.plot(x2,y2)
plt2.title('Robos segun provincias')
plt2.xlabel('provincias')
plt2.ylabel('cantidad de robos y recuperos')
plt2.xticks(rotation=90,fontsize=12)
plt2.show()

print("Cantidad de robos totales: " + data["tramite_tipo"].value_counts())
```

### Respuestas:
* Las marcas Volkswagen, Toyota y Suzuki tienen una mayor cantidad de robos, lo cual puede significar que o que hay mas cantidad que el resto de marcas o que son mas faciles de robar. La provincia de mendoza tiene muchos robos.
* Informacion mas avanzada sobre la seguridad del automovil, como modelo y marca de su alarma, para asi poder plantear si el funcionamiento de alguna de estas es un factor influyente en los robos.
* Dado que la mayoria de los robos y recuperaciones son en mendoza y las marcas mas robadas son Suzuki, Volkswagen y Toyota, especulo que hay una concentracion de esas marcas ahi.
